## 1) Please Enter a Symbol
This block is used to get the stock symbols from the user and then generates a Pandas Df from the data. It then displays the head so we know it worked succesfully, and it also shows us the tail so we know how much data we have.

You may enter up to 4 stock symbols (lowercase or uppercase). If you wish to enter less than 4 simply type 'no' to terminate the prompt.

```python
#This block will loop through and ask the user for stock symbols they want to analyze. 'get_stocks' will promt the user for the symbols and return a list of them.
#'make_df' will make the dataframes based on the symbols, and return a list of dataframes. The for loop will simply print the first row of each DF so that we can confirm it worked.

symbols = bk.get_stocks() 
stock_dfs = bk.make_df(symbols)
stock_dfs_length = len(stock_dfs)
for symbol in range(0,stock_dfs_length):
    print(symbols[symbol])
    display(stock_dfs[symbol].head(1))
    display(stock_dfs[symbol].tail(1))
    bk.line()
```

## 2) Modifies the DF we made to include new columns
Creates new columns taylored to our analysis. The columns include; Fluctuation, Volatility, % Term Moving Average, 5, 50 and 200 day Moving Averages. We also break each days date into differnt columns for the day, month, and year.

```python
#this loops through and modifies each DF for the symbols that we need. 

for symbol in range(0,stock_dfs_length):
    df = stock_dfs[symbol]
    symbol = symbols[symbol]
    print(symbol)
    df= bk.mod_df(df)
    display(df.head(5))
    display(df.tail(1))
    bk.line()
```

## 3) Error Analysis

This section determines the RMSE value for each Moving Average case that we previously predicted to test the accuracy of the moving avergae prediction models

```python
for symbol in range(0,stock_dfs_length):
    df = stock_dfs[symbol]
    symbol = symbols[symbol]
    print(symbol)
    bk.predict(df)
    bk.line()
```

## 4) Todays Summary 
Provides a summary of todays performance for the stock symbols that were entered.

```python
for symbol in range(0,stock_dfs_length):
    print(symbols[symbol])
    bk.todays_summary(stock_dfs[symbol])
    bk.line()
```

## 5) Descriptive Visuals

This section is for showing plots of our data and doing a visual approach to better understand it. These graphs include avg Closing Price per month, avg closing price by day of week, closing price over time, fluctuation over time, and a scatter plot to show the correlation between volume and volatiltiy. We also calcualte the correlation value for the volume and volatitlty as a raw number as well. 

```python
#generates various plots to look at for each symbol. 
for symbol in range(0,stock_dfs_length):
    x = symbol
    df = stock_dfs[symbol]
    symbol = symbols[symbol]
    print(symbol)
    bk.plot_price_vs_month(df, symbol,x)
    bk.plot_price_vs_day(df, symbol,x)
    bk.plot_price_vs_time(df, symbol,x)
    bk.plot_fluc_vs_time(df, symbol,x)
    bk.plot_volume_vs_volatiltiy(df, symbol,x)
    bk.line()
```
